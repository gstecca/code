```python
import networkx as nx
import pandas as pd
```

```python
instance = 'S6'
filename = 'I1_' + instance + '.xlsx'
df = pd.read_excel('instances/' + filename, sheet_name='nodes', index_col='i')
df
```

```python
sol = [0,1,7,10,4,0]
sol2 = [0,2,3,6,0]
sol3 = [0,8,9,5,0]
```

```python
G = nx.Graph()
for index, row in df.iterrows():
    G.add_node(index, pos=(row['x'], row['y']))

for j in range(1, len(sol)):
    print(sol[j-1],sol[j])
    G.add_edge(sol[j-1],sol[j],color='tab:red')

for j in range(1, len(sol2)):
    print(sol2[j-1],sol2[j])
    G.add_edge(sol2[j-1],sol2[j],color='tab:blue')

for j in range(1, len(sol3)):
    print(sol3[j-1],sol3[j])
    G.add_edge(sol3[j-1],sol3[j],color='tab:green')

```

```python
edges = G.edges()
colors = [G[u][v]['color'] for u,v in edges]
nx.draw(G, pos=nx.get_node_attributes(G, 'pos'), with_labels=True, node_size=200 , node_color = 'lightgreen') #
nx.draw_networkx_edges(G, pos=nx.get_node_attributes(G, 'pos'), edge_color=colors, width=2, arrows=True)
#, with_labels=True, node_size=200, linewidths= 0.5, node_color = 'lightgreen'
#G.nodes() edgecolors = 'red', edgecolors = 'red',
```
